```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def rect(a,b,x,y,z,lamb):
    kx = 2*np.pi*x/(lamb*z)
    ky = 2*np.pi*y/(lamb*z)
    return (a*b/(2*np.pi))*np.sinc(a*kx/2)*np.sinc(b*ky/2)
```

```python
lamb = 632e-9 #in m
a = 2e-6 #in m
b = 10e-6 #in m
z = 1 #in m
x = np.linspace(-a/2,a/2,1000)
y = np.linspace(-b/2,b/2,1000)
```

```python
x_p,y_p = np.meshgrid(x,y)
sinc = rect(a,b,x_p,y_p,z,lamb)
```

```python
plt.figure(figsize=(15,10))
plt.contourf(x_p,y_p,sinc, cmap = "jet")
plt.title("Diffraction pattern for a rectangular aperature")
plt.xlabel("x (in m)")
plt.ylabel("y (in m)")
plt.colorbar()
#plt.ylim(-0.01,0.01)
plt.xlim(-0.00001,0.00001)
plt.show()
```
