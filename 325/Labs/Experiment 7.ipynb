```python
import numpy as np
import matplotlib.pyplot as plt
```

#### Michelson Rotating Mirror Method

```python
#data taken from the first cycle of data collection
freq_1 = [50.3,101.3,153.2,198.3,252.5,301.4,350.7,400.0,449.3,498.7,550.3,601.2]
pos_1 = [0,0.85,1.4,1.6,2.1,2.8,3.2,3.8,4.2,4.5,4.9,5.4]
```

```python
#data taken from the second cycle of data collection
freq_2 = [51.4,102.3,151.7,202.6,249.1,301.2,352.2,398.1,449.7,502.8,550.8,600.6]
pos_2 = [0,0.4,0.7,1,1.5,1.8,2.1,2.6,3.2,3.5,4,4.6]
```

```python
#data taken from the third cycle of data collection
freq_3 = [48.4,100.6,151.6,200.5,252.6,300.6,352.2,401.2,452.6,498.1,552.5,603.8]
pos_3 = [0,0.5,1.2,1.6,2.2,2.8,3.3,3.7,4.0,4.4,5.0,5.2]
```

```python
#converting the lists of data into arrays
freq_1 = np.array(freq_1)
freq_2 = np.array(freq_2)
freq_3 = np.array(freq_3)
pos_1 = np.array(pos_1)
pos_2 = np.array(pos_2)
pos_3 = np.array(pos_3)
```

```python
#converting data from mm to m
pos_1 = pos_1 * (1/1000)
pos_2 = pos_2 * (1/1000)
pos_3 = pos_3 * (1/1000)
```

Note that the units of the frequency measurements is Hz, while the units of the displacement of the incoming light is in m

```python
plt.figure(figsize = (12,12))
plt.plot(freq_1,pos_1,'b.',label = "data 1")
plt.plot(freq_2,pos_2,'r.',label = "data 2")
plt.plot(freq_3,pos_3,'g.',label = "data 3")
plt.legend()
plt.grid()
plt.xlabel("Frequency (Hz)")
plt.ylabel("Displacement of light (m)")
plt.title("Light displacement as a function rotating mirror frequency")
```

```python
freq_av = (freq_1 + freq_2 + freq_3) / 3
pos_av = (pos_1 + pos_2 + pos_3) / 3
```

Note that the units of freq_av and pos_av are Hz and m, respectively

```python
#applying a linear fit to the data
poly, Var = np.polyfit(freq_av,pos_av,1, cov = True)
xp = np.linspace(-10,610,1000)
```

```python
Var
```

```python
#finding the error in the slope
#the error is given by the pivot values in the variance matrices
slope_err = np.sqrt(Var[0][0])
slope_err
```

```python
#finding the error of the intercept
#same principle as above
int_err = np.sqrt(Var[1][1])
int_err
```

Note that the respective errors have units of m/Hz and m

```python
poly
```

```python
#applying the fit to the equation y=mx+b
yp = poly[0] * xp + poly[1]
```

```python
#plotting results
plt.figure(figsize = (12,12))
plt.plot(freq_av,pos_av,'go',label = "average data")
plt.plot(xp,yp,label = "fit")
plt.legend()
plt.grid()
plt.xlabel("Frequency (Hz)")
plt.ylabel("Displacement of light (m)")
plt.title("Average light displacement as a function average rotating mirror frequency")
```

```python
print("The slope of the plot is approximately (", round(poly[0],9),"+/-",round(slope_err,8), ") m/Hz")
```

#### Time of Flight Method

```python
time = [107e-9,152e-9,340e-9]
distance = [32.634,46.184,98.77]
```

Note that the units of time are in seconds, while the distance is in meters

```python
plt.figure(figsize = (12,12))
plt.plot(time,distance,'bo',label = "velocity")
plt.grid()
plt.xlabel("Time (s)")
plt.ylabel("Distance (m)")
plt.title("The time it takes light to travel varying distances")
plt.legend()
```

```python
#applying a linear fit to the data
poly, Var = np.polyfit(time,distance,1, cov = True)
xp = np.linspace(1e-7,3.6e-7,1000)
```

```python
poly
```

```python
Var
```

```python
#finding the error in the slope
slope_err = np.sqrt(Var[0][0])
slope_err
```

```python
#applying the fit to the equation y=mx+b
yp = poly[0] * xp + poly[1]
```

Note that the slope and its error are in units of m/s

```python
plt.figure(figsize = (12,12))
plt.plot(time,distance,'go',label = "velocity data")
plt.plot(xp,yp,label = "fit")
plt.legend()
plt.grid()
plt.xlabel("Time (s)")
plt.ylabel("Distance of light travel (m)")
plt.title("Light speed at varying distances")
```

```python
print("The experimentally determined speed of light is approximately (", poly[0],"+/-",slope_err, ") m/s")
```
