```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#constants
L = 6.4 #cm
w = 0.63 #cm
t = 0.16 #cm
lamb = 650e-7 #cm
E = 2.08e12 #dyne/cm^2
m1 = 2.22 #in g
m2 = 5.26 #in g
g = 981 #cm/s^2
```

```python
#The x position's of the bright fringes along the cantilever held 
#down by the smaller weight (in cm)
x_pos1 = [0,2.423,3.253,3.819,4.293,4.741,5.070,5.373,5.662,5.847,6.071,6.295]
#The x position's of the bright fringes along the cantilever held 
#down by the heavier weight (in cm)
x_pos2 = [0,1.987,2.260,2.534,2.772,2.986,3.212,3.402,3.628,3.782,3.937,4.116,4.294,4.425,4.604,
          4.746,4.889,5.032,5.186,5.305,5.424,5.543,5.662,5.805,5.924,6.043,6.281]
```

```python
#formula for calculating the experimental deflection
def exp_defl(N,lamb,alpha,beta):
    num = N * lamb
    denom = np.cos(alpha) + np.cos(beta)
    return num / denom
```

```python
#formula for calculating the theoretical deflection
def theo_defl(F,E,w,t,L,x):
    outside = 12 * F / (E * w * t ** 3)
    inside = (L ** 3 / 6) - ((L-x) **3 / 6) - ((x * L ** 2) / 2)
    return outside * inside
```

```python
#arrays for the number of bright fringes
N1 = []
N2 = []
#appending the number of bright fringes for each case into a list
for i in range(len(x_pos1)):
    N1.append(i)
for j in range(len(x_pos2)):
    N2.append(j)
```

```python
#force of gravity pulling downwards
F1 = -m1 * g
F2 = -m2 * g
```

```python
#the first weight's experimental and theoretical deflections
y1_exp = []
y1_theo = []
for i in N1:
    y1_exp.append(exp_defl(i,lamb,0,0))
for j in x_pos1:
    y1_theo.append(theo_defl(F1,E,w,t,L,j))
```

```python
print(y1_exp)
print()
print(y1_theo)
```

```python
#the second weight's experimental and theoretical deflections
y2_exp = []
y2_theo = []
for i in N2:
    y2_exp.append(exp_defl(i,lamb,0,0))
for j in x_pos2:
    y2_theo.append(theo_defl(F2,E,w,t,L,j))
```

```python
print(y2_exp)
print()
print(y2_theo)
```

```python
plt.figure(figsize = (10,8))
plt.plot(x_pos1,y1_exp,'y.',label = "Experimental deflection")
plt.plot(x_pos1,y1_theo,'b.',label = "Theoretical deflection")
plt.legend()
plt.grid()
plt.xlabel("Position (cm)")
plt.ylabel("Deflection (cm)")
plt.title("Light deflection using the lighter weight")
```

```python
plt.figure(figsize = (10,8))
plt.plot(x_pos2,y2_exp,'g.',label = "Experimental deflection")
plt.plot(x_pos2,y2_theo,'y.',label = "Theoretical deflection")
plt.legend()
plt.grid()
plt.xlabel("Position (cm)")
plt.ylabel("Deflection (cm)")
plt.title("Light deflection using the heavier weight")
```
