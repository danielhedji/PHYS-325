```python
import numpy as np
import matplotlib.pyplot as plt
```

### Preliminaries

```python
#inputting the data from a text file
p1 = np.loadtxt("daniel_data_part1.txt",dtype = str,usecols = (0,1))
p2 = np.loadtxt("daniel_data_part2.txt",dtype = str,usecols = (0,1))
p3 = np.loadtxt("daniel_data_part3.txt",dtype = str ,usecols = (0,1))
```

```python
#since the first point was skewed in the file, I manually changed the first angle to 0 degrees
p1[0,0] = 0
#extracting the part 1 analyzer angle data
p1_ang = p1[:,0]
p1_ang = p1_ang.astype(np.float)
p1_ang
```

Note: Since each of the angles were the same in each part, only p1 was converted to floats to save time. 
Also note that the units of p1_ang (the analyzer angle) are degrees.

```python
#extracting the part 1 irradiance data
p1_I = p1[:,1]
#converting the data to floats
p1_I = p1_I.astype(np.float)
p1_I
```

```python
#extracting the part 2 irradiance data
p2_I = p2[:,1]
#converting the data to floats
p2_I = p2_I.astype(np.float)
p2_I
```

```python
#extracting the part 3 irradiance data
p3_I = p3[:,1]
#converting the data to floats
p3_I = p3_I.astype(np.float)
p3_I
```

Note: the units for p1_I, p2_I, and p3_I are all in lux

```python
#normalizing this data
p1_I = p1_I / max(p1_I)
p1_I
```

p1_I is unitless due to the normalization

### The Law of Malus

```python
#taking the cos of the degree array, first converting the degrees to radians
cos_al = np.cos(np.deg2rad(p1_ang))
#squaring the results
cos_al = np.square(cos_al)
cos_al
```

cos_al is unitless

```python
#organizing the data for every 90 degrees
angs_1 = cos_al[:9]
angs_2 = cos_al[9:18]
angs_3 = cos_al[18:27]
angs_4 = cos_al[27:36]
lum_1 = p1_I[:9]
lum_2 = p1_I[9:18]
lum_3 = p1_I[18:27]
lum_4 = p1_I[27:36]
```

```python
#applying a linear fit to each portion 
poly1 = np.polyfit(angs_1,lum_1,1)
poly2 = np.polyfit(angs_2,lum_2,1)
poly3 = np.polyfit(angs_3,lum_3,1)
poly4 = np.polyfit(angs_4,lum_4,1)
```

```python
#fitting a linear slope to the data
xp = np.linspace(0,1,500)
yp1 = poly1[0] * xp + poly1[1]
yp2 = poly2[0] * xp + poly2[1]
yp3 = poly3[0] * xp + poly3[1]
yp4 = poly4[0] * xp + poly4[1]
```

```python
#data without the fits
plt.figure(figsize = (10,10))
plt.plot(cos_al,p1_I,'b.', label = "Ratio")
plt.legend()
plt.xlabel("cos squared of analyzer angle")
plt.ylabel("Normalized irradiance")
plt.title("Irradiance as a function of analyzer angle")
plt.grid()
```

```python
plt.figure(figsize = (10,10))
plt.plot(cos_al,p1_I,'b.', label = "Ratio")
plt.plot(xp,yp1, label = "fit 1")
plt.plot(xp,yp2, label = "fit 2")
plt.plot(xp,yp3, label = "fit 3")
plt.plot(xp,yp4, label = "fit 4")
plt.legend()
plt.xlabel("cos squared of analyzer angle")
plt.ylabel("Normalized irradiance")
plt.title("Irradiance as a function of analyzer angle")
plt.grid()
```

Note: The extrapolated data has the same units as the plotted points (unitless), and the average
of all four of the slopes is taken to compare to the theoretical slope of 1

```python
print("Slope 1 is equal to", yp1[0])
print("Slope 2 is equal to", yp2[0])
print("Slope 3 is equal to", yp3[0])
print("Slope 4 is equal to", yp4[0])
slope = (abs(yp1[0]) + abs(yp2[0]) + abs(yp3[0]) + abs(yp4[0])) / 4
print("The average slope of the data is", slope)
```

### Retardance

#### Quarter-wave retarder

```python
#plotting the data on a polar plot, with additional graphing options
plt.figure(figsize = (15,10))
ax = plt.subplot(111, projection='polar')
ax.plot(np.deg2rad(p1_ang),p2_I,'.',label = "transmitted irradiance")
plt.legend()
ax.set_rlabel_position(-30)
ax.set_xlabel("Analyzer angle (in rad)")
ax.set_ylabel("Irradiance (in lux)")
plt.title("Irradiance against analyzer angle")
```

By observing the shortest and longest diameter spots on the created plot, the semi-major and minor
axes were able to be approximated.

```python
#finding the semi minor axis diameter
minor_1 = p2_I[13]
minor_2 = p2_I[-7]
semi_minor = abs(minor_1) + abs(minor_2)
print("The semi minor axis value was found to be",semi_minor, "lux")
```

```python
#finding the semi major axis diameter
major_1 = p2_I[2]
major_2 = p2_I[21]
semi_major = abs(major_1) + abs(major_2)
print("The semi major axis value was found to be",semi_major, "lux")
```

```python
asp_rat = semi_major/semi_minor
print("The aspect ratio was found to be", asp_rat)
```
